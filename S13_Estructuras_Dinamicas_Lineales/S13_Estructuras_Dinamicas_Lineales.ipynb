## Fundamentos

- Cola FIFO: primero en entrar, primero en salir.
- collections.deque para colas eficientes en Python.
- Criterios de selección según el comportamiento, no sólo la sintaxis.

## Práctica guiada en Python

Ejecute, analice y modifique el siguiente ejemplo.

```python
from collections import deque

productos = ["Arroz", "Leche"]

pila_deshacer = []
pila_deshacer.append("actualizar P01")
ultima_operacion = pila_deshacer.pop()

cola_pedidos = deque(["V001", "V002"])
cola_pedidos.append("V003")
primer_pedido = cola_pedidos.popleft()

print(productos, ultima_operacion, primer_pedido)
```
