## Práctica guiada en Python

Ejecute, analice y modifique el siguiente ejemplo.

```python
import csv
from pathlib import Path

ARCHIVO = Path("productos.csv")

def cargar_productos(ruta=ARCHIVO):
    if not ruta.exists():
        return []
    with ruta.open(encoding="utf-8", newline="") as archivo:
        return [
            {"codigo": fila["codigo"], "nombre": fila["nombre"], "precio": float(fila["precio"])}
            for fila in csv.DictReader(archivo)
        ]

def guardar_productos(productos, ruta=ARCHIVO):
    with ruta.open("w", encoding="utf-8", newline="") as archivo:
        campos = ["codigo", "nombre", "precio"]
        escritor = csv.DictWriter(archivo, fieldnames=campos)
        escritor.writeheader()
        escritor.writerows(productos)

# El CRUD se realiza sobre esta lista, no directamente sobre el archivo.
productos = cargar_productos()
```
