## Base conceptual de la sesión

**Refactorizar** significa mejorar la estructura interna sin cambiar el comportamiento observable. Si un algoritmo realiza trabajo innecesario, también puede refactorizarse para reducir recorridos o comparaciones.

**Cohesión**: cada función concentra una responsabilidad. **Acoplamiento**: una función no debe depender de detalles innecesarios de otras partes.

La complejidad se analiza de manera introductoria:

- Un recorrido completo crece aproximadamente como `n`.
- Dos ciclos anidados sobre la misma colección crecen aproximadamente como `n²`.
- Detener una búsqueda al encontrar el dato evita trabajo innecesario, aunque el peor caso siga recorriendo la colección.

La persistencia sigue este flujo:

```text
archivo -> carga inicial -> lista en memoria -> CRUD -> guardar instantánea completa -> archivo
```


## Laboratorio de refactorización y optimización

Localiza código repetido, funciones con varias responsabilidades y recorridos que puedan unificarse. Compara una búsqueda que continúa después de encontrar el registro con otra que retorna inmediatamente.


````python
def buscar_indice(datos, codigo):
    for i, elemento in enumerate(datos):
        if elemento["codigo"] == codigo:
            return i
    return -1


def resumir_precios(datos):
    if not datos:
        return 0, 0
    total = 0
    for elemento in datos:
        total += elemento["precio"]
    return total, total / len(datos)
````

## Actividad practica

Refactoriza el CRUD de S9 e incorpora una instantánea para la entidad de tu proyecto.

El programa debe permitir:

1. Cargar registros desde archivo.
2. Registrar un nuevo registro.
3. Listar registros.
4. Buscar por codigo.
5. Editar un registro existente.
6. Eliminar un registro.
7. Guardar los cambios en el archivo.


````python
ARCHIVO = "productos.txt"

productos = []


def cargar():
    productos.clear()
    try:
        with open(ARCHIVO, "r", encoding="utf-8") as archivo:
            for linea in archivo:
                datos = linea.strip().split(";")
                if len(datos) == 4:
                    producto = {
                        "codigo": datos[0],
                        "nombre": datos[1],
                        "precio": float(datos[2]),
                        "stock": int(datos[3])
                    }
                    productos.append(producto)
        print("Datos cargados.")
    except FileNotFoundError:
        print("Archivo no encontrado. Se iniciara con una lista vacia.")


def guardar():
    with open(ARCHIVO, "w", encoding="utf-8") as archivo:
        for p in productos:
            linea = f"{p['codigo']};{p['nombre']};{p['precio']};{p['stock']}\n"
            archivo.write(linea)
    print("Datos guardados.")


def buscar(codigo):
    for p in productos:
        if p["codigo"] == codigo:
            return p
    return None


def registrar():
    codigo = input("Codigo: ")
    if buscar(codigo) is not None:
        print("Ya existe un producto con ese codigo.")
        return

    nombre = input("Nombre: ")
    precio = float(input("Precio: "))
    stock = int(input("Stock: "))

    productos.append({
        "codigo": codigo,
        "nombre": nombre,
        "precio": precio,
        "stock": stock
    })
    print("Producto registrado.")


def listar():
    if len(productos) == 0:
        print("No hay productos registrados.")
        return

    for p in productos:
        print(p["codigo"], p["nombre"], p["precio"], p["stock"])


def editar():
    codigo = input("Codigo a editar: ")
    producto = buscar(codigo)

    if producto is None:
        print("Producto no encontrado.")
        return

    producto["nombre"] = input("Nuevo nombre: ")
    producto["precio"] = float(input("Nuevo precio: "))
    producto["stock"] = int(input("Nuevo stock: "))
    print("Producto actualizado.")


def eliminar():
    codigo = input("Codigo a eliminar: ")
    producto = buscar(codigo)

    if producto is None:
        print("Producto no encontrado.")
        return

    productos.remove(producto)
    print("Producto eliminado.")


def menu():
    cargar()

    while True:
        print("\n1. Registrar")
        print("2. Listar")
        print("3. Buscar")
        print("4. Editar")
        print("5. Eliminar")
        print("6. Guardar")
        print("7. Salir")

        opcion = input("Opcion: ")

        if opcion == "1":
            registrar()
        elif opcion == "2":
            listar()
        elif opcion == "3":
            codigo = input("Codigo a buscar: ")
            producto = buscar(codigo)
            print(producto if producto is not None else "No encontrado")
        elif opcion == "4":
            editar()
        elif opcion == "5":
            eliminar()
        elif opcion == "6":
            guardar()
        elif opcion == "7":
            print("Fin del programa.")
            break
        else:
            print("Opcion no valida.")


menu()
````
