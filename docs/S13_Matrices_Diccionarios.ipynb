## Base conceptual de la sesion

Una **matriz** organiza informacion en filas y columnas. Es util cuando se necesita mostrar o procesar datos tabulares.

Un **diccionario** relaciona una clave con un valor. Es util para consultar por codigo, categoria o cualquier identificador.

En esta sesion no se reemplaza el CRUD: se agregan estructuras de apoyo para procesar mejor la informacion del proyecto.


## Actividad practica

Agrega dos componentes al producto:

1. Un reporte tabular usando una matriz.
2. Un indice o acumulador usando diccionario.

Ejemplos posibles:

- ventas por dia y categoria,
- productos por categoria,
- total vendido por vendedor,
- busqueda rapida por codigo.


```python
productos = [
    {"codigo": "P001", "nombre": "Arroz", "categoria": "Abarrotes", "precio": 4.5, "stock": 20},
    {"codigo": "P002", "nombre": "Azucar", "categoria": "Abarrotes", "precio": 3.8, "stock": 8},
    {"codigo": "P003", "nombre": "Aceite", "categoria": "Cocina", "precio": 9.5, "stock": 5}
]

# Diccionario como indice por codigo
indice = {}
for producto in productos:
    indice[producto["codigo"]] = producto

print("Consulta por codigo:")
print(indice.get("P002", "No encontrado"))

# Diccionario como acumulador por categoria
total_por_categoria = {}
for producto in productos:
    categoria = producto["categoria"]
    valor = producto["precio"] * producto["stock"]

    if categoria not in total_por_categoria:
        total_por_categoria[categoria] = 0

    total_por_categoria[categoria] += valor

print("\nTotal por categoria:")
for categoria, total in total_por_categoria.items():
    print(categoria, total)

# Matriz para reporte tabular
reporte = []
for producto in productos:
    fila = [
        producto["codigo"],
        producto["nombre"],
        producto["categoria"],
        producto["precio"] * producto["stock"]
    ]
    reporte.append(fila)

print("\nReporte tabular:")
for fila in reporte:
    print(fila)
```
