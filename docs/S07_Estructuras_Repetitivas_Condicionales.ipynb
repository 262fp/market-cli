## Base conceptual de la sesión

**Menú interactivo** permite que el usuario elija acciones.

**Validación de opción** evita que una opción inválida rompa el programa.


## Actividad práctica

Integra el registro y listado de S6 dentro de un menú interactivo.

### Arquitectura

```text
Menú -> Opción -> Acción -> Volver al menú -> Salir
```

### Opciones mínimas

1. Registrar
2. Listar
0. Salir


````python
nombres = []
precios = []
stocks = []

opcion = -1
while opcion != 0:
    print('\nMENU PRINCIPAL')
    print('1. Registrar producto')
    print('2. Listar productos')
    print('0. Salir')

    opcion = int(input('Opción: '))

    if opcion == 1:
        nombre = input('Nombre: ')
        precio = float(input('Precio: '))
        stock = int(input('Stock: '))
        nombres.append(nombre)
        precios.append(precio)
        stocks.append(stock)
        print('Producto registrado.')
    elif opcion == 2:
        print('\nLISTADO')
        for i in range(len(nombres)):
            print(i + 1, nombres[i], precios[i], stocks[i])
    elif opcion == 0:
        print('Fin del programa.')
    else:
        print('Opción inválida.')
````
