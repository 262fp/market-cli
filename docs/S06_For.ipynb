## Base conceptual de la sesión

**Estructura lineal**

Una estructura lineal permite guardar varios datos en una secuencia. En Python usaremos listas.

```python
nombres = []
precios = []
stocks = []
```

**Recorrido con for**

`for` permite repetir una acción una cantidad conocida de veces o recorrer los elementos de una lista.

**CRUD inicial**

En esta sesión se implementan las dos primeras operaciones: registrar y listar.


## Actividad práctica

Construye el primer almacenamiento en memoria del proyecto.

### Arquitectura

```text
Entrada -> Registro en listas -> Recorrido con for -> Salida
```

### Pasos

1. Define los campos principales de tu entidad.
2. Crea una lista por cada campo.
3. Registra varios elementos con `append()`.
4. Recorre las listas con `for`.
5. Muestra los registros en formato ordenado.


````python
# CRUD inicial para productos usando listas paralelas
nombres = []
precios = []
stocks = []

cantidad = int(input('Cantidad de productos a registrar: '))

for i in range(cantidad):
    print('Producto', i + 1)
    nombre = input('Nombre: ')
    precio = float(input('Precio: '))
    stock = int(input('Stock: '))

    nombres.append(nombre)
    precios.append(precio)
    stocks.append(stock)

print('\nLISTADO DE PRODUCTOS')
for i in range(len(nombres)):
    print(i + 1, nombres[i], 'Precio:', precios[i], 'Stock:', stocks[i])
````
