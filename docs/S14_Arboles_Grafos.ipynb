## 1. Árboles

Un árbol organiza nodos jerárquicamente. Tiene una raíz, relaciones padre-hijo y hojas sin descendientes. Una jerarquía de categorías es un ejemplo natural.


```python
categorias = {
    "Tienda": ["Abarrotes", "Bebidas"],
    "Abarrotes": ["Arroz", "Azúcar"],
    "Bebidas": ["Agua", "Leche"],
}


def mostrar_arbol(nodo, nivel=0):
    print("  " * nivel + nodo)
    for hijo in categorias.get(nodo, []):
        mostrar_arbol(hijo, nivel + 1)


mostrar_arbol("Tienda")
```

## 2. Recorrido en profundidad

El recorrido anterior es DFS: explora una rama antes de continuar con la siguiente. La recursividad encaja porque cada hijo puede ser tratado como la raíz de un subárbol.


```python
def contiene(nodo, buscado):
    if nodo == buscado:
        return True
    for hijo in categorias.get(nodo, []):
        if contiene(hijo, buscado):
            return True
    return False


assert contiene("Tienda", "Leche")
assert not contiene("Tienda", "Laptop")
```

## 3. Grafos

Un grafo representa una red. Cada vértice puede conectarse con varios otros y pueden existir ciclos. Una lista de adyacencia registra los vecinos de cada vértice.


```python
rutas = {
    "Almacén": ["Caja", "Despacho"],
    "Caja": ["Salida"],
    "Despacho": ["Salida"],
    "Salida": [],
}

for origen, destinos in rutas.items():
    print(origen, "->", destinos)
```

## 4. Recorrido en amplitud

BFS utiliza una cola y visita primero los vecinos cercanos. El conjunto `visitados` evita procesar repetidamente un vértice cuando existen varios caminos o ciclos.


```python
from collections import deque


def recorrido_bfs(grafo, inicio):
    pendientes = deque([inicio])
    visitados = set()
    orden = []
    while pendientes:
        actual = pendientes.popleft()
        if actual in visitados:
            continue
        visitados.add(actual)
        orden.append(actual)
        pendientes.extend(grafo.get(actual, []))
    return orden


assert recorrido_bfs(rutas, "Almacén")[0] == "Almacén"
print(recorrido_bfs(rutas, "Almacén"))
```
