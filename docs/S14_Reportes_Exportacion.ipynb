### Reportes mínimos

1. Exportar registros a CSV.
2. Generar un reporte de texto con encabezado, detalle y totales.
3. Preparar una salida imprimible del reporte o comprobante.


```python
productos = [
    {'codigo': 'P001', 'nombre': 'Arroz', 'precio': 4.5, 'stock': 20},
    {'codigo': 'P002', 'nombre': 'Aceite', 'precio': 9.9, 'stock': 8}
]


def exportar_csv():
    with open('reporte_productos.csv', 'w', encoding='utf-8') as archivo:
        archivo.write('codigo;nombre;precio;stock;importe\n')
        for p in productos:
            importe = p['precio'] * p['stock']
            archivo.write(f"{p['codigo']};{p['nombre']};{p['precio']};{p['stock']};{importe}\n")


def generar_reporte_txt():
    total_general = 0
    with open('reporte_productos.txt', 'w', encoding='utf-8') as archivo:
        archivo.write('REPORTE DE PRODUCTOS\n')
        archivo.write('=' * 40 + '\n')
        for p in productos:
            importe = p['precio'] * p['stock']
            total_general += importe
            archivo.write(f"{p['codigo']} - {p['nombre']} - Importe: {importe}\n")
        archivo.write('=' * 40 + '\n')
        archivo.write(f'TOTAL GENERAL: {total_general}\n')

exportar_csv()
generar_reporte_txt()
```
