## 1. Lista dinámica

Una lista permite agregar, consultar, actualizar y eliminar elementos. Es adecuada cuando se necesita recorrer o acceder a diferentes posiciones.


```python
productos = []
productos.append({"codigo": "P001", "nombre": "Arroz"})
productos.append({"codigo": "P002", "nombre": "Leche"})

productos[1]["nombre"] = "Leche evaporada"
for producto in productos:
    print(producto)
```

## 2. Pila: comportamiento LIFO

Una pila atiende primero el último elemento agregado. Puede representar historial para deshacer, navegación o procesamiento pendiente en profundidad.


```python
pila_deshacer = []
pila_deshacer.append(("actualizar", "P001", 4.5))
pila_deshacer.append(("eliminar", "P002"))

ultima_operacion = pila_deshacer.pop()
print("Deshacer:", ultima_operacion)
assert ultima_operacion[0] == "eliminar"
```

## 3. Cola: comportamiento FIFO

Una cola atiende primero el elemento que llegó primero. `deque` permite extraer eficientemente desde el extremo izquierdo.


```python
from collections import deque

cola_pedidos = deque()
cola_pedidos.append("V001")
cola_pedidos.append("V002")
cola_pedidos.append("V003")

primero = cola_pedidos.popleft()
print("Atendiendo:", primero)
assert primero == "V001"
```

## 4. Comparación y selección

| Necesidad | Estructura |
|---|---|
| Consultar y modificar elementos | Lista |
| Deshacer la última operación | Pila |
| Atender solicitudes por llegada | Cola |

No se debe incorporar una estructura sólo para cumplir una lista de temas. La elección se fundamenta en el comportamiento requerido.


```python
from collections import deque

def atender_todos(pendientes):
    atendidos = []
    while pendientes:
        atendidos.append(pendientes.popleft())
    return atendidos

assert atender_todos(deque(["A", "B", "C"])) == ["A", "B", "C"]
```
