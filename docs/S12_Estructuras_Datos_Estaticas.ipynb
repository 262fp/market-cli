## 1. Matrices bidimensionales

En Python una matriz se representa mediante una lista que contiene otras listas. Cada fila puede representar un producto y cada columna un período.

```text
             Día 1  Día 2  Día 3
Producto 1      3      2      4
Producto 2      1      5      2
```


````python
productos = ["Arroz", "Leche", "Azúcar"]
ventas = [
    [3, 2, 4],
    [1, 5, 2],
    [4, 3, 1],
]

for fila in range(len(ventas)):
    total = 0
    for columna in range(len(ventas[fila])):
        total += ventas[fila][columna]
    print(productos[fila], "total vendido:", total)
````

## 2. Registro, consulta y capacidad lógica

Aunque las listas de Python son dinámicas, el ejercicio puede trabajar con una capacidad definida para comprender índices y límites. Es importante validar fila y columna antes de acceder.


````python
FILAS = 3
COLUMNAS = 4
cantidades = [[0 for _ in range(COLUMNAS)] for _ in range(FILAS)]

def registrar_cantidad(fila, columna, cantidad):
    if not (0 <= fila < FILAS and 0 <= columna < COLUMNAS):
        return False
    if cantidad < 0:
        return False
    cantidades[fila][columna] = cantidad
    return True

assert registrar_cantidad(0, 1, 5)
assert not registrar_cantidad(5, 1, 3)
````

## 3. Arreglos tridimensionales

Una tercera dimensión puede representar locales, almacenes o ciclos. El orden de las dimensiones debe documentarse para evitar confusiones.


````python
# local -> producto -> día
ventas_por_local = [
    [[3, 2], [1, 5]],
    [[2, 4], [3, 1]],
]

for local, matriz in enumerate(ventas_por_local, start=1):
    total_local = 0
    for fila in matriz:
        for cantidad in fila:
            total_local += cantidad
    print("Local", local, "total:", total_local)
````

## 4. Arreglos paralelos

Los arreglos paralelos relacionan datos mediante el mismo índice. Son útiles para comprender correspondencia y capacidad, pero cualquier inserción, eliminación u ordenamiento debe aplicarse a todos los arreglos.


````python
codigos = ["P001", "P002", "P003"]
nombres = ["Arroz", "Leche", "Azúcar"]
precios = [4.5, 4.2, 3.8]

for i in range(len(codigos)):
    print(codigos[i], nombres[i], precios[i])

assert len(codigos) == len(nombres) == len(precios)
````
