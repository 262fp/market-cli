## Base conceptual de la sesión

Una **función** agrupa instrucciones que resuelven una tarea concreta. Los parámetros representan sus entradas y `return` comunica el resultado sin mezclarlo con la presentación.

```python
def calcular_subtotal(precio, cantidad):
    return precio * cantidad
```

Una función cohesiva hace una sola cosa. En TDD se escribe primero un caso esperado, se implementa la solución mínima y luego se mejora el código conservando su comportamiento.

```text
Prueba que falla -> solución mínima -> prueba que pasa -> refactorización
```

TDD puede emplearse desde las condicionales: no está limitado a funciones, pero las funciones pequeñas facilitan aislar y automatizar los casos.


## Actividad práctica

Refactoriza una solución secuencial y una condicional de sesiones anteriores.

### Funciones mínimas

- `calcular_subtotal(precio, cantidad)`
- `calcular_descuento(subtotal)`
- `validar_cantidad(cantidad, stock)`
- `calcular_total(precio, cantidad)`

Antes de completar cada función, define casos normales, límites e inválidos.


````python
def calcular_subtotal(precio, cantidad):
    if precio < 0 or cantidad <= 0:
        raise ValueError("Precio y cantidad deben ser válidos")
    return precio * cantidad


def calcular_descuento(subtotal):
    if subtotal >= 300:
        return subtotal * 0.10
    return 0.0


def calcular_total(precio, cantidad):
    subtotal = calcular_subtotal(precio, cantidad)
    return subtotal - calcular_descuento(subtotal)
````

## TDD con pruebas pequeñas

Las pruebas expresan el comportamiento antes de modificar la implementación. Se recomienda comenzar con un caso sencillo, agregar el límite y finalmente validar entradas incorrectas.


````python
import unittest


class PruebasVenta(unittest.TestCase):
    def test_subtotal(self):
        self.assertEqual(30, calcular_subtotal(10, 3))

    def test_descuento_desde_limite(self):
        self.assertEqual(30, calcular_descuento(300))

    def test_total_con_descuento(self):
        self.assertEqual(270, calcular_total(100, 3))

    def test_cantidad_invalida(self):
        with self.assertRaises(ValueError):
            calcular_subtotal(10, 0)


suite = unittest.defaultTestLoader.loadTestsFromTestCase(PruebasVenta)
unittest.TextTestRunner(verbosity=2).run(suite)
````
