# Sesion 10: Archivos y CRUD persistente

## Objetivo
Implementar persistencia basica con archivos para que el CRUD del proyecto pueda guardar y cargar registros.

## Producto de la sesion
CRUD persistente en archivos: registrar, listar, buscar, editar y eliminar registros, guardando y cargando informacion desde archivo.

## Resultado esperado
La aplicacion carga registros desde un archivo al iniciar, trabaja con una lista en memoria y guarda los cambios en el archivo cuando el usuario lo solicita o al salir.

## Alcance de la sesion
- Archivo de texto o CSV como almacenamiento simple.
- Carga inicial de registros.
- Guardado completo de la lista en memoria.
- Registro, listado, busqueda, edicion y eliminacion.
- Separacion entre datos en memoria y datos persistidos.

## Idea fuerza
No se trabaja con base de datos. La persistencia se resuelve con archivos para consolidar listas, funciones, busqueda, edicion y eliminacion antes de pasar a reportes.


## Base conceptual de la sesion

**Persistencia** significa que los datos no se pierden cuando termina el programa.

En esta sesion se usa una estrategia simple:

```text
archivo -> cargar registros -> lista en memoria -> registrar/editar/eliminar -> guardar lista completa -> archivo
```

Para el curso, el archivo completo se reescribe al guardar. Esta decision es suficiente porque el volumen de datos es pequeno y permite entender claramente el flujo de lectura, procesamiento y escritura.


## Actividad practica

Construye un CRUD persistente para la entidad de tu proyecto.

El programa debe permitir:

1. Cargar registros desde archivo.
2. Registrar un nuevo registro.
3. Listar registros.
4. Buscar por codigo.
5. Editar un registro existente.
6. Eliminar un registro.
7. Guardar los cambios en el archivo.


````python
ARCHIVO = "productos.txt"

productos = []


def cargar():
    productos.clear()
    try:
        with open(ARCHIVO, "r", encoding="utf-8") as archivo:
            for linea in archivo:
                datos = linea.strip().split(";")
                if len(datos) == 4:
                    producto = {
                        "codigo": datos[0],
                        "nombre": datos[1],
                        "precio": float(datos[2]),
                        "stock": int(datos[3])
                    }
                    productos.append(producto)
        print("Datos cargados.")
    except FileNotFoundError:
        print("Archivo no encontrado. Se iniciara con una lista vacia.")


def guardar():
    with open(ARCHIVO, "w", encoding="utf-8") as archivo:
        for p in productos:
            linea = f"{p['codigo']};{p['nombre']};{p['precio']};{p['stock']}\n"
            archivo.write(linea)
    print("Datos guardados.")


def buscar(codigo):
    for p in productos:
        if p["codigo"] == codigo:
            return p
    return None


def registrar():
    codigo = input("Codigo: ")
    if buscar(codigo) is not None:
        print("Ya existe un producto con ese codigo.")
        return

    nombre = input("Nombre: ")
    precio = float(input("Precio: "))
    stock = int(input("Stock: "))

    productos.append({
        "codigo": codigo,
        "nombre": nombre,
        "precio": precio,
        "stock": stock
    })
    print("Producto registrado.")


def listar():
    if len(productos) == 0:
        print("No hay productos registrados.")
        return

    for p in productos:
        print(p["codigo"], p["nombre"], p["precio"], p["stock"])


def editar():
    codigo = input("Codigo a editar: ")
    producto = buscar(codigo)

    if producto is None:
        print("Producto no encontrado.")
        return

    producto["nombre"] = input("Nuevo nombre: ")
    producto["precio"] = float(input("Nuevo precio: "))
    producto["stock"] = int(input("Nuevo stock: "))
    print("Producto actualizado.")


def eliminar():
    codigo = input("Codigo a eliminar: ")
    producto = buscar(codigo)

    if producto is None:
        print("Producto no encontrado.")
        return

    productos.remove(producto)
    print("Producto eliminado.")


def menu():
    cargar()

    while True:
        print("\n1. Registrar")
        print("2. Listar")
        print("3. Buscar")
        print("4. Editar")
        print("5. Eliminar")
        print("6. Guardar")
        print("7. Salir")

        opcion = input("Opcion: ")

        if opcion == "1":
            registrar()
        elif opcion == "2":
            listar()
        elif opcion == "3":
            codigo = input("Codigo a buscar: ")
            producto = buscar(codigo)
            print(producto if producto is not None else "No encontrado")
        elif opcion == "4":
            editar()
        elif opcion == "5":
            eliminar()
        elif opcion == "6":
            guardar()
        elif opcion == "7":
            guardar()
            print("Fin del programa.")
            break
        else:
            print("Opcion no valida.")


menu()
````
