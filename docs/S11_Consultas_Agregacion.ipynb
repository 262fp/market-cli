### Consultas mínimas

- Listar registros filtrados.
- Contar registros.
- Sumar un campo numérico.
- Calcular promedio.
- Obtener mínimo y máximo.


```python
productos = [
    {'codigo': 'P001', 'nombre': 'Arroz', 'precio': 4.5, 'stock': 20, 'categoria': 'abarrotes'},
    {'codigo': 'P002', 'nombre': 'Aceite', 'precio': 9.9, 'stock': 8, 'categoria': 'abarrotes'},
    {'codigo': 'P003', 'nombre': 'Jabón', 'precio': 3.5, 'stock': 15, 'categoria': 'limpieza'}
]

cantidad = len(productos)
stock_total = 0
precio_total = 0
precio_min = productos[0]['precio']
precio_max = productos[0]['precio']

for p in productos:
    stock_total += p['stock']
    precio_total += p['precio']
    if p['precio'] < precio_min:
        precio_min = p['precio']
    if p['precio'] > precio_max:
        precio_max = p['precio']

print('Cantidad:', cantidad)
print('Stock total:', stock_total)
print('Precio promedio:', round(precio_total / cantidad, 2))
print('Precio mínimo:', precio_min)
print('Precio máximo:', precio_max)
```

```python
# Agrupación básica por categoría
conteo = {}

for p in productos:
    categoria = p['categoria']
    if categoria not in conteo:
        conteo[categoria] = 0
    conteo[categoria] += 1

print(conteo)
```
