## Práctica guiada en Python

Ejecute, analice y modifique el siguiente ejemplo.

```python
ventas = [120.0, 85.5, 210.0, 64.5]
total = 0.0
mayores_a_cien = 0

for posicion, venta in enumerate(ventas, start=1):
    total += venta
    if venta > 100:
        mayores_a_cien += 1
    print(posicion, venta)

print("Total:", total)
print("Mayores a 100:", mayores_a_cien)
```
