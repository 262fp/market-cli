
**CSV o texto delimitado**

Un formato simple para guardar registros separando campos por `;`, `,` u otro símbolo.


## Actividad práctica

Agrega persistencia al proyecto.

### Arquitectura

```text
Archivo -> Cargar datos -> Trabajar en memoria -> Guardar datos -> Archivo
```

### Funciones mínimas

- `cargar_datos()`
- `guardar_datos()`
- `registrar()`
- `listar()`


````python
ARCHIVO = 'productos.csv'
productos = []


def cargar_datos():
    productos.clear()
    try:
        with open(ARCHIVO, 'r', encoding='utf-8') as archivo:
            for linea in archivo:
                codigo, nombre, precio, stock = linea.strip().split(';')
                productos.append({
                    'codigo': codigo,
                    'nombre': nombre,
                    'precio': float(precio),
                    'stock': int(stock)
                })
    except FileNotFoundError:
        pass


def guardar_datos():
    with open(ARCHIVO, 'w', encoding='utf-8') as archivo:
        for p in productos:
            archivo.write(f"{p['codigo']};{p['nombre']};{p['precio']};{p['stock']}\n")


def registrar():
    codigo = input('Código: ')
    nombre = input('Nombre: ')
    precio = float(input('Precio: '))
    stock = int(input('Stock: '))
    productos.append({'codigo': codigo, 'nombre': nombre, 'precio': precio, 'stock': stock})
    guardar_datos()


def listar():
    for p in productos:
        print(p)

cargar_datos()
````
