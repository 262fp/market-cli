# Sesión 9: Operaciones con arreglos

## Objetivo
Implementar un CRUD completo sobre los registros almacenados en memoria, reutilizando la búsqueda y controlando duplicados e inexistencias.

## Producto de la sesión
CRUD completo en memoria: registrar, listar, consultar, actualizar y eliminar registros.

## Resultado esperado
Gestiona una colección durante la ejecución del programa y mantiene consistentes sus datos sin utilizar todavía archivos.

## Alcance de la sesión
- Registro y listado.
- Consulta mediante búsqueda por código.
- Actualización y eliminación.
- Control de duplicados y registros inexistentes.
- Separación entre menú y operaciones.


## Base conceptual de la sesión

El CRUD reúne cuatro operaciones: crear, consultar, actualizar y eliminar. En esta sesión todas trabajan únicamente sobre la lista en memoria.

```text
menú -> función CRUD -> lista en memoria -> resultado
```

La función de búsqueda devuelve el índice o el registro encontrado. Este resultado evita repetir recorridos y permite validar duplicados antes de registrar.


## 1. Datos y búsqueda reutilizable

Se utilizará una lista de diccionarios para mantener unidos los campos de cada producto.


````python
productos = [
    {"codigo": "P001", "nombre": "Arroz", "precio": 4.5, "stock": 20},
    {"codigo": "P002", "nombre": "Leche", "precio": 4.2, "stock": 10},
]


def buscar_indice(codigo):
    for i, producto in enumerate(productos):
        if producto["codigo"].lower() == codigo.lower():
            return i
    return -1


assert buscar_indice("P001") == 0
assert buscar_indice("P999") == -1
````

## 2. Registrar y listar

Registrar debe rechazar códigos repetidos. Listar no modifica la colección y debe responder claramente cuando está vacía.


````python
def registrar(producto):
    if buscar_indice(producto["codigo"]) != -1:
        return False
    productos.append(producto)
    return True


def listar():
    if not productos:
        print("No hay productos registrados")
        return
    for producto in productos:
        print(producto["codigo"], producto["nombre"], producto["precio"], producto["stock"])


assert registrar({"codigo": "P003", "nombre": "Azúcar", "precio": 3.8, "stock": 8})
assert not registrar({"codigo": "P003", "nombre": "Duplicado", "precio": 1, "stock": 1})
````

## 3. Consultar, actualizar y eliminar

Las operaciones devuelven `False` cuando el código no existe; el menú decide qué mensaje presentar.


````python
def consultar(codigo):
    indice = buscar_indice(codigo)
    return None if indice == -1 else productos[indice]


def actualizar(codigo, nombre, precio, stock):
    indice = buscar_indice(codigo)
    if indice == -1:
        return False
    productos[indice].update(nombre=nombre, precio=precio, stock=stock)
    return True


def eliminar(codigo):
    indice = buscar_indice(codigo)
    if indice == -1:
        return False
    productos.pop(indice)
    return True


assert actualizar("P001", "Arroz superior", 5.0, 18)
assert consultar("P001")["stock"] == 18
assert not eliminar("P999")
````
