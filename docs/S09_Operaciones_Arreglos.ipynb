## Práctica guiada en Python

Ejecute, analice y modifique el siguiente ejemplo.

```python
productos = [
    {"codigo": "P01", "nombre": "Arroz", "precio": 4.8},
    {"codigo": "P02", "nombre": "Leche", "precio": 4.2},
]

def buscar_indice(datos, codigo):
    for i, producto in enumerate(datos):
        if producto["codigo"] == codigo:
            return i
    return -1

def registrar(datos, producto):
    if buscar_indice(datos, producto["codigo"]) != -1:
        return False
    datos.append(producto)
    return True

def actualizar_precio(datos, codigo, precio):
    indice = buscar_indice(datos, codigo)
    if indice == -1:
        return False
    datos[indice]["precio"] = precio
    return True

def eliminar(datos, codigo):
    indice = buscar_indice(datos, codigo)
    if indice == -1:
        return False
    datos.pop(indice)
    return True
```
