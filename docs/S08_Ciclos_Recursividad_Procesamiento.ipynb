## Base conceptual de la sesión

La **búsqueda lineal** recorre los elementos hasta encontrar una coincidencia. Un **ordenamiento elemental** compara e intercambia valores y normalmente emplea ciclos anidados.

Los acumulados resumen una colección: total, promedio, máximo y mínimo. Una función **recursiva** se llama a sí misma y siempre necesita un caso base alcanzable.

Al trabajar con listas paralelas, cualquier intercambio debe conservar alineados todos los datos del registro.

```python
nombres = ['Arroz', 'Azucar', 'Aceite']
precios = [4.5, 3.8, 9.9]
stocks = [20, 15, 8]

buscado = input('Nombre a buscar: ')
encontrado = False

for i in range(len(nombres)):
    if nombres[i].lower() == buscado.lower():
        print('Encontrado:', nombres[i], precios[i], stocks[i])
        encontrado = True

if not encontrado:
    print('No encontrado')
```

```python
# Ordenación básica por nombre manteniendo listas paralelas
for i in range(len(nombres) - 1):
    for j in range(i + 1, len(nombres)):
        if nombres[i] > nombres[j]:
            nombres[i], nombres[j] = nombres[j], nombres[i]
            precios[i], precios[j] = precios[j], precios[i]
            stocks[i], stocks[j] = stocks[j], stocks[i]

for i in range(len(nombres)):
    print(nombres[i], precios[i], stocks[i])
```

```python
def resumir(valores):
    if not valores:
        return {"total": 0, "promedio": 0, "maximo": None, "minimo": None}

    total = 0
    maximo = valores[0]
    minimo = valores[0]
    for valor in valores:
        total += valor
        if valor > maximo:
            maximo = valor
        if valor < minimo:
            minimo = valor
    return {
        "total": total,
        "promedio": total / len(valores),
        "maximo": maximo,
        "minimo": minimo,
    }


print(resumir([80, 35, 120, 65]))
```

```python
def suma_recursiva(valores, posicion=0):
    if posicion == len(valores):
        return 0
    return valores[posicion] + suma_recursiva(valores, posicion + 1)


assert suma_recursiva([10, 20, 30]) == 60
assert suma_recursiva([]) == 0
```
