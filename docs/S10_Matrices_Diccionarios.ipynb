## Base conceptual de la sesión

```python

```

**Diccionario**

Relaciona una clave con un valor.

```python

```


## Actividad práctica

Desarrolla dos componentes:

1. Un reporte tabular usando matriz.
2. Un registro o consulta usando diccionario.


````python
# Matriz: ventas por producto y día
ventas = [
    [5, 3, 4],
    [2, 6, 1],
    [7, 2, 5]
]

for i in range(len(ventas)):
    total = 0
    for j in range(len(ventas[i])):
        total += ventas[i][j]
    print('Producto', i + 1, 'total vendido:', total)
````

````python
# Diccionario: registro por código
productos = {
    'P001': {'nombre': 'Arroz', 'precio': 4.5, 'stock': 20},
    'P002': {'nombre': 'Aceite', 'precio': 9.9, 'stock': 8}
}

codigo = input('Código a consultar: ')

if codigo in productos:
    print(productos[codigo])
else:
    print('No existe')
````
