## Fundamentos

- Ciclo while, condición de continuidad y centinelas.
- Validación repetitiva de datos.
- Menús controlados por opción.
- Python no posee do-while; el patrón while True con break permite ejecutar antes de comprobar la salida.
- Prevención de ciclos infinitos.

## Práctica guiada en Python

Ejecute, analice y modifique el siguiente ejemplo.

```python
productos = []

while True:
    print("1. Registrar  2. Listar  0. Salir")
    opcion = input("Opción: ").strip()
    if opcion == "1":
        nombre = input("Nombre: ").strip()
        if nombre:
            productos.append(nombre)
    elif opcion == "2":
        print(productos)
    elif opcion == "0":
        break
    else:
        print("Opción inválida")
```
