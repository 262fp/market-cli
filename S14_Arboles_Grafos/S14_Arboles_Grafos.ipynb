## Práctica guiada en Python

Ejecute, analice y modifique el siguiente ejemplo.

```python
from collections import deque

categorias = {
    "Tienda": ["Abarrotes", "Bebidas"],
    "Abarrotes": ["Arroz", "Azúcar"],
    "Bebidas": ["Agua"],
}

def recorrer_arbol(nodo, nivel=0):
    print("  " * nivel + nodo)
    for hijo in categorias.get(nodo, []):
        recorrer_arbol(hijo, nivel + 1)

rutas = {"Almacén": ["Caja"], "Caja": ["Salida"], "Salida": []}
pendientes = deque(["Almacén"])
visitados = set()
while pendientes:
    actual = pendientes.popleft()
    if actual not in visitados:
        visitados.add(actual)
        pendientes.extend(rutas[actual])

recorrer_arbol("Tienda")
print(visitados)
```
