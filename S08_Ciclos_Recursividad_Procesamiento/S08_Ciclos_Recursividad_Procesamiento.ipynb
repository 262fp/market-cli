## Práctica guiada en Python

Ejecute, analice y modifique el siguiente ejemplo.

```python
def ordenar_burbuja(datos):
    resultado = datos.copy()
    for limite in range(len(resultado) - 1, 0, -1):
        for i in range(limite):
            if resultado[i] > resultado[i + 1]:
                resultado[i], resultado[i + 1] = resultado[i + 1], resultado[i]
    return resultado

def suma_recursiva(datos):
    if not datos:
        return 0
    return datos[0] + suma_recursiva(datos[1:])

ventas = [80, 35, 120, 65]
total = suma_recursiva(ventas)
print(ordenar_burbuja(ventas))
print(total, total / len(ventas), min(ventas), max(ventas))
```
