## Práctica guiada en Python

Ejecute, analice y modifique el siguiente ejemplo.

```python
def calcular_subtotal(precio, cantidad):
    if precio < 0 or cantidad <= 0:
        raise ValueError("Precio y cantidad deben ser válidos")
    return precio * cantidad

def aplicar_descuento(subtotal):
    return subtotal * 0.90 if subtotal >= 300 else subtotal

# Pruebas escritas antes o junto con la implementación.
assert calcular_subtotal(10, 3) == 30
assert aplicar_descuento(300) == 270

try:
    calcular_subtotal(10, 0)
    raise AssertionError("Se esperaba ValueError")
except ValueError:
    pass
```
