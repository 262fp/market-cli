## Base conceptual de la sesion

Una **consulta** recupera registros que cumplen una condicion.

Una **agregacion** resume varios registros en un solo resultado, por ejemplo: cantidad, suma, promedio, minimo o maximo.

Un **reporte** presenta informacion procesada de forma clara para revisar el estado del proyecto.


## Actividad practica

Agrega reportes al CRUD persistente de la sesion anterior.

Consultas minimas:

- Listar productos con stock bajo.
- Contar registros.
- Sumar el valor total del inventario.
- Calcular precio promedio.
- Obtener el producto con mayor y menor precio.


```python
productos = [
    {"codigo": "P001", "nombre": "Arroz", "precio": 4.5, "stock": 20},
    {"codigo": "P002", "nombre": "Azucar", "precio": 3.8, "stock": 8},
    {"codigo": "P003", "nombre": "Aceite", "precio": 9.5, "stock": 5}
]


def contar_productos():
    return len(productos)


def valor_inventario():
    total = 0
    for p in productos:
        total += p["precio"] * p["stock"]
    return total


def precio_promedio():
    if len(productos) == 0:
        return 0

    suma = 0
    for p in productos:
        suma += p["precio"]
    return suma / len(productos)


def productos_stock_bajo(limite):
    resultado = []
    for p in productos:
        if p["stock"] <= limite:
            resultado.append(p)
    return resultado


def producto_mayor_precio():
    if len(productos) == 0:
        return None

    mayor = productos[0]
    for p in productos:
        if p["precio"] > mayor["precio"]:
            mayor = p
    return mayor


def reporte():
    print("Cantidad de productos:", contar_productos())
    print("Valor total del inventario:", valor_inventario())
    print("Precio promedio:", precio_promedio())
    print("Producto de mayor precio:", producto_mayor_precio())
    print("Productos con stock bajo:")
    for p in productos_stock_bajo(10):
        print(p)


reporte()
```
